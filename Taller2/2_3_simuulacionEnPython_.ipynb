2.3 ✅ Demuestre si los siguientes sistemas de la forma  y=Hx , son sistemas lineales e invariantes en el tiempo (SLIT) (simule los sistemas en Python):

* \\( y[n] = \frac{x[n]}{3} + 2x[n-1] - y[n-1] \\)\

```python
import numpy as np
np.random.seed(0)
N=200; x1=np.random.randn(N); x2=np.random.randn(N); a,b=1.7,-0.6
def H1(x):
    y=np.zeros_like(x)
    for n in range(N):
        y[n]=(1/3)*x[n] + 2*(x[n-1] if n>0 else 0) - (y[n-1] if n>0 else 0)
    return y
def shift(v,k): return np.r_[np.zeros(k), v[:-k]] if k>0 else np.r_[v[-k:], np.zeros(-k)]
lin = np.allclose(H1(a*x1+b*x2), a*H1(x1)+b*H1(x2))
TI  = np.allclose(H1(shift(x1,5)), shift(H1(x1),5))
print("S1 -> lineal:",lin," TI:",TI)
```

*   \\( y[n] = \sum_{k=-\infty}^{n} x^2[k] \\)

```python
import numpy as np

# --- Sistema 2 ---
def H2(x):
    """ y[n] = sum_{k=-inf}^{n} x[k]^2 """
    return np.cumsum(x.astype(float)**2)

# Corrimiento causal con ceros
def shift(v, k):
    if k >= 0:
        return np.r_[np.zeros(k), v[:-k]]
    else:
        k = -k
        return np.r_[v[k:], np.zeros(k)]

# ---- PRUEBA DE PROPIEDADES ----
np.random.seed(1)
N = 200
x1 = np.random.randn(N)
x2 = np.random.randn(N)
a, b = 1.7, -0.6

# Prueba de linealidad
lin = np.allclose(H2(a*x1 + b*x2), a*H2(x1) + b*H2(x2))

# Prueba de TI (ignorando los primeros valores por el corrimiento)
n0 = 5
H_shift = H2(shift(x1, n0))
shifted_H = shift(H2(x1), n0)
TI = np.allclose(H_shift[10:], shifted_H[10:])

print("S2 -> lineal:", lin, " TI:", TI)
```

*   \\( y[n] = \text{median}(x[n]) \\); donde median es la función mediana sobre una ventana de tamaño 3.

```python
import numpy as np

# --- Sistema 3: Filtro de mediana con ventana 3 ---
def H3(x):
    x = x.astype(float)
    N = len(x)
    y = np.zeros_like(x)
    # Para evitar errores de borde, solo se calcula entre 1 y N-2
    for n in range(1, N-1):
        ventana = [x[n-1], x[n], x[n+1]]
        y[n] = np.median(ventana)
    return y

# Función de corrimiento (shift) causal con ceros
def shift(v, k):
    if k >= 0:
        return np.r_[np.zeros(k), v[:-k]]
    else:
        k = -k
        return np.r_[v[k:], np.zeros(k)]

# ---- PRUEBA DE PROPIEDADES ----
np.random.seed(2)
N = 200
x1 = np.random.randn(N)
x2 = np.random.randn(N)
a, b = 1.7, -0.6

# Prueba de linealidad
lin = np.allclose(H3(a*x1 + b*x2), a*H3(x1) + b*H3(x2))

# Prueba de TI (ignorando bordes para no afectar resultado)
n0 = 5
H_shift = H3(shift(x1, n0))
shifted_H = shift(H3(x1), n0)

# Solo comparamos desde la muestra 10 hasta N-10 (para evitar bordes)
TI = np.allclose(H_shift[10:-10], shifted_H[10:-10])

print("S3 -> lineal:", lin, " TI:", TI)
```

*   \\( y(t) = Ax(t) + B \\); con \\( A, B \in \mathbb{R} \\)

```python
import numpy as np
np.random.seed(3)
N=200; x1=np.random.randn(N); x2=np.random.randn(N); a,b=1.7,-0.6
A,B=2.5,0.3   # pon B=0 para que sea lineal
def H4(x): return A*x + B
def shift(v,k): return np.r_[np.zeros(k), v[:-k]] if k>0 else np.r_[v[-k:], np.zeros(-k)]
lin = np.allclose(H4(a*x1+b*x2), a*H4(x1)+b*H4(x2))   # False si B!=0, True si B=0
TI  = np.allclose(H4(shift(x1,5)), shift(H4(x1),5))   # True
print("S4 -> lineal:",lin," TI:",TI)
```
